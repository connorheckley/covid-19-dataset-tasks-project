```python
data.groupby('Region').sum().head(20)
# Showing the number of confirmed cases, confirmed deaths and recovered cases in each region.
# head(20) will show first 20 countries within the dataset.
```

```python
data.groupby('Region')['Confirmed'].sum()
# Only shows the number of confirmed cases per country.
```

```python
data.groupby('Region')['Confirmed'].sum() .sort_values(ascending = True)
# This shows the dataset sorted by highest number of confirmed cases at the top.
# Alternatively if (ascending = True) is used it will display the region with the lowest number of confirmed cases at the top.
```

```python
data.groupby('Region')['Deaths'].sum() .sort_values(ascending = False)
# This shows the dataset sorted by highest number of deaths at the top.
# Alternatively if (ascending = True) is used it will display the region with the lowest number of deaths at the top.
```

```python
data.groupby('Region')['Recovered'].sum() .sort_values(ascending = False)
# This displays the country with the highest number of patients recovered.
# Once again, if (ascending = False) is changed to (ascending = True) then the country with the lowest recovered patients will be displayed at the top.
```

```python
data.groupby('Region')['Recovered'].sum() .sort_values(ascending = False).head(15)
# This displays the first 15 results in the dataset for this query.
# The number can be changed to anything within the maximum value in the dataset and 'Recovered' can be changed to 'Confirmed' and 'Deaths'.
# ascending can also be set to True or False.
```

```python
data.groupby('Region')['Confirmed', 'Recovered'].sum()
# Displaying both the columns for Confirmed and Recovered cases.
```

```python
data[data.Confirmed < 10]
# This shows all places where Confirmed cases are less than 10
# The small arrow can be changed to show Confirmed cases that are greater than a number (> 10)
```

```python
data[data.Recovered < 10]
# This shows all places where Recovered cases are less than 10
# The small arrow can be changed to show recovered cases that are greater than a number (> 10)
```

```python
data[data.Deaths < 10]
# This shows all places where Deaths are less than 10
# The small arrow can be changed to show deaths that are greater than a number (> 10)
```

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending = False)
# Displays the region with the most confirmed cases and the region with the lowest confirmed cases in ascending order.
```

```python
data.groupby('Region').Deaths.sum().sort_values(ascending = True)
# Displays the region with the lowest death count.
```

```python
data.groupby('Region').Deaths.sum().sort_values(ascending = True).head(10)
# The .head(10) function only displays the first 10 results and can be changed to any number.
```

```python
data[data.Region == 'France']
# This shows the results from only one country/entry in the dataset.
# Will only work if that word is contained within the dataset.
```

```python
data.sort_values( by = ['Confirmed'] , ascending = True)
# Shows the total number of confirmed cases in ascending order.
```

```python
data.sort_values( by = ['Recovered'] , ascending = False)
# Shows the number of Recovered cases in descending order
```
